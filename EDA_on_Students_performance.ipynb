## Importing the necessary libraries and CSV file

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Load the dataset
df = pd.read_csv('/content/StudentsPerformance.csv')
df.head()
```

## Question 1: How are the scores distributed across the different subjects?

```python
# Plot the distribution of scores for each subject
plt.hist(df['math score'], alpha=0.5, label='Math')
plt.hist(df['reading score'], alpha=0.5, label='Reading')
plt.hist(df['writing score'], alpha=0.5, label='Writing')
plt.legend(loc='upper right')
plt.title('Distribution of Scores')
plt.xlabel('Score (out of 100)')
plt.ylabel('Frequency')
plt.show()
```

## **Observations:**

The distribution of scores for each subject appears to be roughly normal, with a peak around the middle of the range and decreasing frequency as the scores approach the high and low ends of the range.
The reading and writing scores appear to be more similar to each other than either is to the math scores.

## Question 2: What is the relationship between test preparation and scores?

```python
import seaborn as sns

# Create a boxplot of scores by test preparation status
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title('Math Scores by Test Preparation Status')
plt.xlabel('Test Preparation Course')
plt.ylabel('Score (out of 100)')
plt.show()

sns.boxplot(x='test preparation course', y='reading score', data=df)
plt.title('Reading Scores by Test Preparation Status')
plt.xlabel('Test Preparation Course')
plt.ylabel('Score (out of 100)')
plt.show()

sns.boxplot(x='test preparation course', y='writing score', data=df)
plt.title('Writing Scores by Test Preparation Status')
plt.xlabel('Test Preparation Course')
plt.ylabel('Score (out of 100)')
plt.show()
```

## **Observations:**

- Students who completed the test preparation course tend to have higher scores than students who did not complete the course.
- The difference in scores between students who completed the course and those who did not appears to be largest for math, followed by writing and reading.
- There is more variability in the scores of students who did not complete the course, especially for math.

## Question 3: Is there a difference in scores based on parental level of education or race/ethnicity? 

```python
# Define the order of the parental level of education groups for plotting
parental_order = ['some high school', 'high school', 'some college', "associate's degree", "bachelor's degree", "master's degree"]

# Define the order of the race/ethnicity groups for plotting
race_order = ['group A', 'group B', 'group C', 'group D', 'group E']

# Create a barplot of average scores by parental level of education
df_edu = df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean().loc[parental_order]
df_edu.plot(kind='bar')
plt.title('Average Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Score (out of 100)')
plt.show()

# Create a barplot of average scores by race/ethnicity
df_race = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean().loc[race_order]
df_race.plot(kind='bar')
plt.title('Average Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Score (out of 100)')
plt.show()
```

## **Observations:**

- There is a clear positive relationship between parental level of education and student scores, with students whose parents have higher levels of education tending to have higher scores than students whose parents have lower levels of education.
- The difference in scores between students with parents at different levels of education appears to be largest for math, followed by reading and writing.
- There are some differences in average scores by race/ethnicity, with Group E tending to have the highest scores and Group A tending to have the lowest scores. However, the differences are relatively small compared to the differences based on parental level of education

## Question 4: Is there a relationship between scores and gender?

```python
# Create a boxplot of scores by gender
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Score (out of 100)')
plt.show()

sns.boxplot(x='gender', y='reading score', data=df)
plt.title('Reading Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Score (out of 100)')
plt.show()

sns.boxplot(x='gender', y='writing score', data=df)
plt.title('Writing Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Score (out of 100)')
plt.show()
```

## Observations:

- There appears to be a small difference in scores between male and female students, with female students tending to have slightly higher scores than male students.
- The difference in scores between male and female students is largest for reading, followed by writing and math.
- There is more variability in the scores of male students than female students for math and writing, but not for reading.
